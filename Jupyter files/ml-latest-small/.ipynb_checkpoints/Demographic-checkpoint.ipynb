```python
'''
Testing different ways of recommending movies
This was not used in final product
'''
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
users = pd.read_csv("users.csv", sep=";")
users

ratings1 = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
ratings = pd.merge(movies, ratings1, on = 'movieId')
information = pd.merge(ratings, users, on='userId')


fem_under18 = information.loc[(information.Occupation == 'writer')&(information.rating >= 3.5)&(information.Age == '35-44')]

# count rows per `movie_id`
#abs_num_f_5s_ratings = f_5s_ratings.groupby("Movie_id").size()

top_movies = fem_under18.groupby("movieId").size()
top_movies
#Merge the two data sets on movieID
top_movies

movie_data = pd.merge(ratings1, movies, on='movieId')  
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False)
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())  

ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())  

fem_under18

fem_avg_rating = pd.DataFrame(fem_under18.groupby('title')['rating'].mean())
#Get count of ammount of votes per movie and add to DF
fem_avg_rating['count_ratings'] = fem_under18.groupby('title')['rating'].count()
#avg_rating

top_ten_rated = fem_avg_rating.sort_values('rating', ascending=False).head(10)
ten_most_voted = fem_avg_rating.sort_values('count_ratings', ascending=False).head(10)


hundred_most_voted = fem_avg_rating.sort_values('count_ratings', ascending=False).head(50)
top_10 = hundred_most_voted.sort_values('rating', ascending=False).head(10)

top_10
```
