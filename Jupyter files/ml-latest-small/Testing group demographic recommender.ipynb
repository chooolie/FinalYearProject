```python
'''
This is the code used for the 
demographic recommender system
'''
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
users = pd.read_csv("users.csv")
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
ratings2 = pd.merge(movies, ratings, on = 'movieId')
information = pd.merge(ratings2, users, on='userId')
links = pd.read_csv("links.csv")

def age_occupation(age,occupation):
    user_demo = information.loc[(information.Occupation == occupation)&(information.rating >= 2.5)&(information.Age == age)]
    movie_data = pd.merge(ratings, movies, on='movieId')  
    #movie_data.groupby('title')['rating'].mean().sort_values(ascending=False)
    ratings_mean_count = pd.DataFrame(movie_data.groupby('movieId')['rating'].mean())  
    #movie_data = movie_data.groupby('title')['rating'].mean()
    #ratings_mean_count = pd.DataFrame(movie_data)
    ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('movieId')['rating'].count())  
    user_avg_rating = pd.DataFrame(user_demo.groupby('movieId')['rating'].mean())
    user_avg_rating['count_ratings'] = user_demo.groupby('movieId')['rating'].count()
    hundred_most_voted = user_avg_rating.sort_values('count_ratings', ascending=False).head(50)
    top_10 = hundred_most_voted.sort_values('rating', ascending=False).head(10)
    sim = pd.merge(top_10, links, on='movieId')
    sim = pd.merge(sim, movies, on='movieId')
    sim = sim.values
    return sim

user1 = age_occupation('Under18', 'K-12student')
#user2 = age_occupation('45-49','scientist')
rec_movies = age_occupation('25-34', 'academic/educator')
movies = []
for mov in rec_movies:
    movies.append(int(mov[4]))
movies
```
