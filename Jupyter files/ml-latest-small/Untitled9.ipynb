```python
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import pandas as pd 
users = pd.read_csv("users.csv")
ratings = pd.read_csv("ratings.csv")
links = pd.read_csv("links.csv")
movies = pd.read_csv("movies.csv")
ratings2 = pd.merge(movies, ratings, on = 'movieId')
information = pd.merge(ratings2, users, on='userId')

def age_occupation(age,occupation):
    user_demo = information.loc[(information.Occupation == occupation)&(information.rating >= 2.5)&(information.Age == age)]
    movie_data = pd.merge(ratings, movies, on='movieId')
    ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
    ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
    user_avg_rating = pd.DataFrame(user_demo.groupby('movieId')['rating'].mean())
    user_avg_rating['count_ratings'] = user_demo.groupby('movieId')['rating'].count()
    hundred_most_voted = user_avg_rating.sort_values('count_ratings', ascending=False).head(50)
    top_10 = hundred_most_voted.sort_values('rating', ascending=False).head(15)
    sim = pd.merge(top_10, links, on='movieId')
    sim = pd.merge(sim, movies, on='movieId')
    sim = sim.values
    return sim

def gender_age(age,gender):
    user_demo = information.loc[(information.Gender == gender)&(information.rating >= 2.5)&(information.Age == age)]
    movie_data = pd.merge(ratings, movies, on='movieId')
    ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
    ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
    user_avg_rating = pd.DataFrame(user_demo.groupby('movieId')['rating'].mean())
    user_avg_rating['count_ratings'] = user_demo.groupby('movieId')['rating'].count()
    hundred_most_voted = user_avg_rating.sort_values('count_ratings', ascending=False).head(50)
    top_10 = hundred_most_voted.sort_values('rating', ascending=False).head(15)
    sim = pd.merge(top_10, links, on='movieId')
    sim = pd.merge(sim, movies, on='movieId')
    sim = sim.values
    return sim



def gender_occupation(gender,occupation):
    user_demo = information.loc[(information.Occupation == occupation)&(information.rating >= 3)&(information.Gender == gender)]
    movie_data = pd.merge(ratings, movies, on='movieId')
    ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
    ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
    user_avg_rating = pd.DataFrame(user_demo.groupby('movieId')['rating'].mean())
    user_avg_rating['count_ratings'] = user_demo.groupby('movieId')['rating'].count()
    hundred_most_voted = user_avg_rating.sort_values('count_ratings', ascending=False).head(50)
    top_10 = hundred_most_voted.sort_values('rating', ascending=False).head(15)
    sim = pd.merge(top_10, links, on='movieId')
    sim = pd.merge(sim, movies, on='movieId')
    sim = sim.values
    return sim
young_student = age_occupation('18-24', 'college/grad student')
young_female = gender_age('18-24', 'F')
female_student = gender_occupation('F', 'college/grad student')

rec_movies = age_occupation('18-24', 'college/grad student')
movies = []
for mov in young_student:
    movies.append(int(mov[4]))

#young_female = gender_age('18-24', 'F')
movies2= []
for mov in young_female:
    movies2.append(int(mov[4]))

#rec_movies3 = gender_occupation('F', 'college/grad student')
movies3= []
for mov in female_student:
    movies3.append(int(mov[4]))


most_frequent = []
most_frequent.extend(movies)
most_frequent.extend(movies2)
most_frequent.extend(movies3)
recommendations = [word for word, word_count in Counter(most_frequent).most_common(10)]
recommendations 
rec = pd.DataFrame({'movieId':recommendations})
rec
new = pd.DataFrame()
m = pd.read_csv("movies.csv")
new = pd.merge(rec, m, on='movieId' )

new
```
